# Inspecting Fortune 500 USA companies 2020-2025

```python
import pandas as pd

companies = set()
for year in range(2020, 2026):
    df = pd.read_csv(f"data/fortune_500/fortune500_USA_{year}.csv")
    companies.update(df["company"])

print("Unique companies between 2020 and 2025: ", len(companies))
```
